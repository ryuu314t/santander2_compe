```python
import numpy as np
import pandas as pd
from sklearn import neighbors
from sklearn.cluster import KMeans
```

```python
%%time
train_df = pd.read_csv("./input/train.csv")
test_df = pd.read_csv("./input/test.csv")
```

```python
submit_flg=1 #全件処理かどうかをコントロール
grid_flg=1 #GridSearchをするかコントロール
SEED=12345
sample_num=10000
fold_num=5

#train関連
train_drop_col=['ID_code', 'target']
train_label='target'

#test関連
test_drop_col=['ID_code']
```

```python
#x,y作成
x_train=train_df.drop(train_drop_col,axis=1)
y_train=train_df[train_label]
x_test=test_df.drop(test_drop_col,axis=1)
```

特徴量作成コードを記載

```python
clf=KMeans(n_clusters=10, init='k-means++', n_init=10, max_iter=300,
                       tol=0.0001,precompute_distances='auto', verbose=0,
                       random_state=None, copy_x=True, n_jobs=1)
```

```python
df_kmeans=pd.concat([x_train,x_test])
pred=clf.fit_predict(df_kmeans)

df_kmeans['label']=pred
```

```python
train_num=len(x_train)
test_num=len(x_test)

df_train_kmeans=df_kmeans['label'].head(train_num)
df_test_kmeans=df_kmeans['label'].tail(test_num)
```

```python
train_df['label']=df_train_kmeans
test_df['label']=df_test_kmeans
```

```python
train_df.to_csv('./feature/102_train_kmeans.csv',index=False)
test_df.to_csv('./feature/102_test_kmeans.csv',index=False)
```
